```python
""" 기간별 생산량 """

import matplotlib.pyplot as plt
import pandas as pd
import math as mt
data = pd.read_csv("과일류_생산량.csv", engine='python')
data.head()
```

```python
data["총 생산량"] = data.sum(axis=1)
data.head()
```

```python
group_names = ['적다','보통','많다']
data['비교 생산량'],mybin = pd.cut(data['총 생산량'],3,labels=group_names,retbins=True)
data
```

```python
csv = pd.concat([data, pd.get_dummies(data['비교 생산량'])],axis=1,sort=False)
csv
```
